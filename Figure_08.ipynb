```python
%matplotlib inline
# This magic command ensures plots appear in the notebook

import numpy as np
import matplotlib.pyplot as plt

# Update global font settings (DejaVu Sans is the default, but we set it explicitly)
plt.rcParams.update({'font.family': 'DejaVu Sans', 'font.size': 12})

# Generate an array of x values from 0 to 10 with a step of 0.1
x = np.arange(0, 10.1, 0.1)

# Compute the sine and cosine values for each x
y1 = np.sin(x)
y2 = np.cos(x)

# Add reproducible random noise (seed=0 to match MATLAB rng(0))
np.random.seed(0)
noise_amplitude = 0.2
y1_noisy = y1 + noise_amplitude * np.random.randn(len(y1))
y2_noisy = y2 + noise_amplitude * np.random.randn(len(y2))

# Create a new figure for the plot
plt.figure(figsize=(7.5, 5))

# Generate two perceptually uniform colors from the 'RdGy' colormap
cmap = plt.get_cmap('RdGy', 2)
color1 = cmap(0)
color2 = cmap(1)

# Plot data with random noise, thicker lines, and distinct line styles
plt.plot(x, y1_noisy, '--', linewidth=2, color=color1, label='Sine Wave + Noise')
plt.plot(x, y2_noisy, '-', linewidth=2, color=color2, label='Cosine Wave + Noise')

# Set the plot title and axis labels with larger fonts for better readability
plt.title('Sine and Cosine Waves', fontsize=16)
plt.xlabel('Time (s)', fontsize=14)
plt.ylabel('Amplitude (mV)', fontsize=14)

# Annotate key points at x = 
idx_closest = np.argmin(np.abs(x - np.pi))
key_x = x[idx_closest]
key_y_sin_noisy = y1_noisy[idx_closest]

# Highlight critical points with markers
plt.scatter(key_x, key_y_sin_noisy, s=75, color='k', label='_nolegend_')

# Define the arrow starting and ending points
arrow_start_x = key_x + 0.6
arrow_start_y = key_y_sin_noisy + 0.6

# Add the arrow to the plot
plt.annotate(
    '',
    xy=(key_x, key_y_sin_noisy),
    xytext=(arrow_start_x, arrow_start_y),
    arrowprops=dict(arrowstyle='->', lw=1.5, color='k', mutation_scale=20),
    annotation_clip=False
)

# Add annotations
plt.text(arrow_start_x, arrow_start_y + 0.1, 'Key Point', fontsize=14)

# Adjust Y-axis limits to be symmetric
plt.ylim([-2, 2])
plt.yticks(np.arange(-2, 3, 1))

# Add grid, box, and legend
plt.grid(True)
ax = plt.gca()
for spine in ax.spines.values():
    spine.set_visible(True)
plt.legend(loc='lower left', fontsize=12)

# Show the figure
plt.show()
```

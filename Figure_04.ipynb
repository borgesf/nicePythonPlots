```python
%matplotlib inline
# This magic command ensures plots appear in the notebook

import numpy as np
import matplotlib.pyplot as plt

# Update global font settings (DejaVu Sans is the default, but we set it explicitly)
plt.rcParams.update({'font.family': 'DejaVu Sans', 'font.size': 12})

# Generate an array of x values from 0 to 10 with a step of 0.1
x = np.arange(0, 10.1, 0.1)

# Compute the sine and cosine values for each x
y1 = np.sin(x)
y2 = np.cos(x)

# Create a new figure for the plot
plt.figure()

# Plot the sine and cosine curves with labels for the legend
plt.plot(x, y1, label='Sine Wave')
plt.plot(x, y2, label='Cosine Wave')

# Set the plot title and axis labels with larger fonts for better readability
plt.title('Sine and Cosine Waves', fontsize=16)
plt.xlabel('Time (s)', fontsize=14)
plt.ylabel('Amplitude (mV)', fontsize=14)

# Define a key point at x =  and compute its sine value
key_x = np.pi
key_y_sin = np.sin(key_x)

# Mark the key point with a black dot (marker size 20)
plt.plot(key_x, key_y_sin, '.', markersize=20, color='k')

# Add a text annotation near the key point with a small vertical offset
plt.text(key_x, key_y_sin + 0.1, 'Key Point', fontsize=12)

# Display the legend in the upper right corner
plt.legend(loc='upper right', fontsize=12)

# Show the plot
plt.show()
```

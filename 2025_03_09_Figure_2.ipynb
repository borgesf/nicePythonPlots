```python
%matplotlib inline
# This ensures plots are displayed within the notebook

import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 10 with a step of 0.1
x = np.arange(0, 10.1, 0.1)

# Compute sine and cosine values for each x
y1 = np.sin(x)
y2 = np.cos(x)

# Create a new figure for the plot
plt.figure()

# Plot the sine and cosine curves with labels for the legend
plt.plot(x, y1, label='Sine Wave')
plt.plot(x, y2, label='Cosine Wave')

# Set the title and axis labels for clarity
plt.title('Sine and Cosine Waves')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (mV)')

# Place the legend in the lower left corner
plt.legend(loc='lower left')

# Display the plot
plt.show()
```

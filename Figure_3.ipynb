```python
%matplotlib inline
# Notebook 3: Customizing Font Type and Font Size with Two Figures

import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 10 with a step of 0.1
x = np.arange(0, 10.1, 0.1)
# Compute sine and cosine values for each x
y1 = np.sin(x)
y2 = np.cos(x)

# -----------------------------
# Figure 1: Default Font Settings
# -----------------------------
plt.figure()
plt.plot(x, y1, label='Sine Wave')
plt.plot(x, y2, label='Cosine Wave')
plt.title('Sine and Cosine Waves (Default Fonts)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (mV)')
plt.legend(loc='lower left')
plt.show()

# -----------------------------
# Figure 2: Customized Font Settings
# -----------------------------
# Update global settings for font family and base font size
plt.rcParams.update({'font.family': 'DejaVu Sans', 'font.size': 12})

plt.figure()
plt.plot(x, y1, label='Sine Wave')
plt.plot(x, y2, label='Cosine Wave')
# Set title and labels with larger fonts for improved readability
plt.title('Sine and Cosine Waves (Custom Fonts)', fontsize=16)
plt.xlabel('Time (s)', fontsize=14)
plt.ylabel('Amplitude (mV)', fontsize=14)
plt.legend(loc='lower left', fontsize=12)
plt.show()
```
